```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Initialize Chrome WebDriver (make sure to download chromedriver.exe and place it in your PATH)
driver = webdriver.Chrome()

# URL of the first page
base_url = "https://www.mrmed.in/products"

lists = []

# Loop through 5 pages
for page in range(1, 5):
    url = base_url.format(page)
    driver.get(url)
    
    # Wait for the page to load (adjust the sleep time as needed)
    time.sleep(3)
    
    # Find elements for name and storage
    names = driver.find_elements(By.XPATH, '//div[@class="ProductCard_mrp__2LCND"]')
    storages = driver.find_elements(By.XPATH, '//div[@class="ProductCard_salt_name__W2S3q"]')
    Med = driver.find_elements(By.XPATH, '//div[@class="ProductCard_product_title__3ZtHe"]')
    
    # Check if lengths match
    if len(names) != len(storages):
        print(f"Error on page {page}: Lengths of names and storages do not match.")
        continue
    
    # Append data to lists
    for i in range(len(names)):
        lists.append([names[i].text, storages[i].text, Med[i].text])
    
# Close the WebDriver
driver.quit()

# Create a DataFrame
df = pd.DataFrame(lists, columns=['Name', 'Storage', 'Med'])

# Print the DataFrame
#print(df)
#split_df = pd.DataFrame(df['Name'].str.split('\n', expand=True))
#split_df.rename(columns={0: 'Name', 1: 'Qty', 2: 'Salt', 3: 'Company'}, inplace = True)
#Price = df[['Storage']].rename(columns = {'Storage':'Price'})
#Data_D = pd.concat([split_df, Price], axis = 1)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Initialize Chrome WebDriver (make sure to download chromedriver.exe and place it in your PATH)
driver = webdriver.Chrome()

# URL of the first page
base_url = "https://www.mrmed.in/products?page={}"

lists = []

# Loop through 5 pages
for page in range(1, 335):  # Changed range to fetch 5 pages
    url = base_url.format(page)
    driver.get(url)
    
    # Wait for the page to load (adjust the sleep time as needed)
    time.sleep(3)
    
    # Find elements for name, storage, and medicine details
    names = driver.find_elements(By.XPATH, '//div[@class="ProductCard_mrp__2LCND"]')
    storages = driver.find_elements(By.XPATH, '//div[@class="ProductCard_salt_name__W2S3q"]')
    meds = driver.find_elements(By.XPATH, '//div[@class="ProductCard_product_title__3ZtHe"]')
    
    # Check if lengths match
    if len(names) != len(storages) or len(names) != len(meds):
        print(f"Error on page {page}: Lengths of names, storages, and meds do not match.")
        continue
    
    # Append data to lists
    for i in range(len(names)):
        lists.append([names[i].text, storages[i].text, meds[i].text])
    
# Close the WebDriver
driver.quit()

# Create a DataFrame
df = pd.DataFrame(lists, columns=['Name', 'Storage', 'Medicine'])

# Print the DataFrame
#print(df.head())  # Print first few rows of the DataFrame
```
