```python
grouped_data = Manual.groupby('id').agg({
    'hospital_code' : lambda x: list(x),
    'routing_reason': lambda x: list(x),
    'review_reasons': lambda x: list(x),
    
}).reset_index()

# Display the grouped data
#print(grouped_data)
```

```python
ids_with_null_reasons = []

# Initialize an empty list to store dictionaries
data_list = []

# Iterate through each row in the DataFrame
for idx, row in grouped_data.iterrows():
    id_value = row['id']
    routing_reasons = row['routing_reason']
    review_reasons = row['review_reasons']
    
    # Combine routing_reason and review_reasons into a single string (reason)
    reasons_list = routing_reasons + review_reasons  # Concatenate lists
    reasons_list = [reason for reason in reasons_list if pd.notna(reason)]  # Remove NaN values
    reason = ', '.join(map(str, reasons_list))  # Convert to string
    
    # Create a dictionary for each row
    data_dict = {
        'id': id_value,
        'reason': reason
    }
    
    # Append the dictionary to the list
    data_list.append(data_dict)
    
    # Check if reason is null or empty, then save the id
    if pd.isna(reason) or reason.strip() == '':
        ids_with_null_reasons.append(id_value)

# Convert the list of dictionaries to a DataFrame
captured_data = pd.DataFrame(data_list)

# Display the captured data
print("Captured Data:")
print(captured_data)

# Display IDs with null or empty reasons
print("\nIDs with Null or Empty Reasons:")
print(ids_with_null_reasons)
```
