```python
import pandas as pd
import re
bills_data = bills_data.astype(str)
database_data = database_data.astype(str)

# Function to escape regex special characters in patterns
def escape_regex(pattern):
    return re.escape(pattern)

# Function to filter database_data based on a single row from bills_data using regex
def filter_database_with_regex(bills_row):
    medicine_pattern = f"^{escape_regex(bills_row['Medicine_Name_Bill'])}$"
    quantity_pattern = f"^{escape_regex(bills_row['Quantity_Bill'])}$"
    dosage_pattern = f"^{escape_regex(bills_row['Dosage_Bill'])}$"
    
    return database_data[
        database_data['Medicine_Name_DB'].str.match(medicine_pattern) &
        database_data['Quantity_DB'].str.match(quantity_pattern) &
        database_data['Dosage_DB'].str.match(dosage_pattern)
    ]

# Apply the filtering function to each row of bills_data
filtered_results = pd.concat([filter_database_with_regex(row) for _, row in bills_data.iterrows()])

# Drop duplicate rows in case of overlaps
filtered_results = filtered_results.drop_duplicates()

# Optionally, you might want to select specific columns
filtered_results = filtered_results[['Medicine_Name_DB', 'Quantity_DB', 'Dosage_DB', 'Database_Rates']]

# Print or save the filtered data
print(filtered_results)
# Optionally save to a CSV
# filtered_results.to_csv('filtered_data.csv', index=False)
```

```python
# Ensure all relevant columns are treated as strings for regex matching
bills_data['Quantity_Bill'] = pd.to_numeric(bills_data['Quantity_Bill'], errors='coerce')
bills_data['Dosage_Bill'] = bills_data['Dosage_Bill'].fillna('')
bills_data['Medicine_Name_Bill'] = bills_data['Medicine_Name_Bill'].fillna('')

database_data['Quantity_DB'] = pd.to_numeric(database_data['Quantity_DB'], errors='coerce')
database_data['Dosage_DB'] = database_data['Dosage_DB'].fillna('')
database_data['Medicine_Name_DB'] = database_data['Medicine_Name_DB'].fillna('')

# Function to filter database_data based on a single row from bills_data
def filter_database_with_regex(bills_row):
    # Construct regex patterns with escaping
    medicine_pattern = f"^{re.escape(str(bills_row['Medicine_Name_Bill']))}$"
    quantity_pattern = f"^{re.escape(str(bills_row['Quantity_Bill']))}$"
    dosage_pattern = f"^{re.escape(str(bills_row['Dosage_Bill']))}$"
    
    # Filter using the patterns
    filtered = database_data[
        database_data['Medicine_Name_DB'].str.match(medicine_pattern) &
        database_data['Quantity_DB'].astype(str).str.match(quantity_pattern) &
        database_data['Dosage_DB'].str.match(dosage_pattern)
    ]
    
    return filtered

# Apply the filtering function to each row of bills_data
filtered_results = pd.concat([filter_database_with_regex(row) for _, row in bills_data.iterrows()])

# Drop duplicate rows in case of overlaps
filtered_results = filtered_results.drop_duplicates()

# Optionally, you might want to select specific columns
filtered_results = filtered_results[['Medicine_Name_DB', 'Quantity_DB', 'Dosage_DB', 'Database_Rates']]

# Print or save the filtered data
print("Final filtered results:")
print(filtered_results)
```
