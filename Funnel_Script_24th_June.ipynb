```python
unique_count_id = data['id'].nunique()

data.drop_duplicates(subset = ['id'], keep = 'first', inplace = True)

Default_hopital_Adjucation_not_enable = data[data['default_hospital']==1]['default_hospital'].count()

df = data[data['default_hospital']!=1]
df.routing_reason.count()

filter_values = ['POLICY_PRODUCT_OUT_OF_SCOPE', 'POLICY_PRODUCT_OUT_OF_SCOPE|This is an reconsideration case.']

filtered_df = df[df['routing_reason'].isin(filter_values)]
Policy_product_not_live=filtered_df['routing_reason'].count()

filtered_policy = df[~df['routing_reason'].isin(filter_values)]
filter_values_others = ['Others','others', 'OTHERS']
others = filtered_policy[filtered_policy['others'].isin(filter_values_others)]
Procedure_Submitted_in_Others= others['others'].count()

filter_others = filtered_policy[~filtered_policy['others'].isin(filter_values_others)]
filter_values_procedure = ['PROCEDURE_OUT_OF_SCOPE',
    'PROCEDURE_OUT_OF_SCOPE|This is an reconsideration case.','PROCEDURE_OUT_OF_SCOPE|BILL_PARSING_ISSUE|BILL_PARSING_ISSUE',
'PROCEDURE_OUT_OF_SCOPE|This is an extension case.','PROCEDURE_OUT_OF_SCOPE|PROVIDER_OUT_OF_SCOPE_DISCHARGE',
                 'PROCEDURE_OUT_OF_SCOPE|This is an reconsideration case.|This is an reconsideration case.']
procedure = filter_others[filter_others['routing_reason'].isin(filter_values_procedure)]
procedure_not_live = procedure['routing_reason'].count()

procedure_filter = filter_others[~filter_others['routing_reason'].isin(filter_values_procedure)]

claim_list = ['CLAIM_CREATED_WITHOUT_MEMBER_VERICATION',    'CLAIM_CREATED_WITHOUT_MEMBER_VERICATION|This is an extension case.', 
              'CLAIM_CREATED_WITHOUT_MEMBER_VERICATION|BILL_PARSING_ISSUE|BILL_PARSING_ISSUE',
             'CLAIM_CREATED_WITHOUT_MEMBER_VERICATION|This is an reconsideration case.']
claim_without = procedure_filter[procedure_filter['routing_reason'].isin(claim_list)]  
Member_Id_not_verified = claim_without.id.count()

claim_filter = procedure_filter[~procedure_filter['routing_reason'].isin(claim_list)]

ported_list = ['This policy is ported.', 'This policy is ported. | Underlying cause is Malignant neoplasm of uterus,part unspecified moving it to manual.']
ported = claim_filter[claim_filter['review_reasons'].isin(ported_list)]
ported_policy = ported['review_reasons'].count()

ported_filter = claim_filter[~claim_filter['review_reasons'].isin(ported_list)]


first_claim = ported_filter[ported_filter['review_reasons'] == 'This is the first claim for the procedure.']
Dialysis_first_claim = first_claim['review_reasons'].count()

first_claim_filter = ported_filter[ported_filter['review_reasons'] != 'This is the first claim for the procedure.']

Auto = first_claim_filter[first_claim_filter['preauth_result']=='A']
Auto_Filter = first_claim_filter[first_claim_filter['preauth_result']!='A']
Auto_ = Auto['preauth_result'].count()


Manual = Auto_Filter
Manual_ = Manual.id.count()
```
