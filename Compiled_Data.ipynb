```python
import pandas as pd
import numpy as np
```

```python
import os
```

```python
excel_folder = r'C:\Users\Vitraya1\Downloads\Tata_mg'
```

```python
# List Excel files in the directory
files = os.listdir(excel_folder)

# Read Excel files using pandas
dfs = []

for file in files:
    if file.endswith('.xlsx') or file.endswith('.xls'):  # Check if the file is Excel
        path = os.path.join(excel_folder, file)
        df = pd.read_excel(path)
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame (optional)
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df
```

```python
# Save to Excel
excel_file = 'Combined_Tata.xlsx'
combined_df.to_excel(excel_file, index=False)

print(f"Data saved to {excel_file}")
```
